```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_meal=pd.read_csv("meal_info.csv")
```

```python
df_meal.head()
```

```python
df_center=pd.read_csv("fulfilment_center_info.csv")
```

```python
df_center.head()
```

```python
df_food=pd.read_csv('train.csv')
```

```python
df_food.head()
```

```python
df_food
```

```python
df_week_rev_n=pd.pivot(df_food,columns="week",values=["num_orders"])
```

```python
df_week_rev_n=df_week_rev_n.fillna(0)
df_week_rev_n=np.array(df_week_rev_n)
shape(df_week_rev_n)
```

```python
df_week_rev_p=pd.pivot(df_food,columns="week",values=["checkout_price"])
```

```python
df_week_rev_p=df_week_rev_p.fillna(0)
df_week_rev_p=np.array(df_week_rev_p)
df_week_rev_p
```

```python
df_rev_w=np.dot(df_week_rev_n.T,df_week_rev_p)
```

```python
df_rev_w=pd.DataFrame(df_rev_w.diagonal())
df_rev_w=df_rev_w.rename(columns={0:"revenue"})
```

```python
df_rev_w=pd.DataFrame.append(pd.DataFrame([0],columns=["revenue"]),df_rev_w,).reset_index().iloc[:,1:]
```

```python
monthly=[]
sum1=0
for i in range(len(df_rev_w)):
    sum1+=df_rev_w.iloc[i,0]
    if i%4==0:
        monthly.append(sum1)
        sum1=0
        
```

```python
df_rev_m=pd.DataFrame(monthly)
```

```python
df_rev_m.columns=["revenue"]
```

```python
plt.plot(df_rev_m,label="monthly")
plt.xlabel('weeks')
plt.ylabel('revenue')
plt.legend()
plt.title('month rev')
plt.show()
```

```python
plt.xlabel("weeks")
plt.ylabel("revenue")
plt.plot(df_rev_w)
plt.title('week rev')
plt.show()
```

```python
df=pd.merge(df_food,df_center)
```

```python
df=pd.merge(df,df_meal)
```

```python
table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)
table
```

```python
table=table.reset_index()
table
```

```python
plt.xlabel='Category'
plt.ylabel='Num_orders'
plt.title("Most Popular food")
plt.xticks(rotation=70)
plt.bar(table.iloc[:,0],table.iloc[:,1])
plt.show()
```
