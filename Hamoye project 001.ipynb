```python
import pandas as pd
import numpy as np
```

```python
cell_phone=pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv')
population=pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv')
geo_country=pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
```

```python
cell_phone
```

```python
df=pd.merge(cell_phone,population,'inner',on=['year','Country'])
#cell_phone.
```

```python
geo_country=geo_country.rename(columns={"country":"geo","name":"Country"})
geo_country
```

```python
df1=df.merge(geo_country[['geo','Country']],how='inner')
df1
```

```python
no_of_cellphone_per_person=df1['cell_phones_total']/df1['population']
```

```python
df1["geo"]=df1["geo"].str.upper()
df1.describe()

```
