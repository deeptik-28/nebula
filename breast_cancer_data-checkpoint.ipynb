```python
# Converting categorical data into numerical data using dictionary
dict={'Deceased':0,'Alive':1}
df['Patient\'s Vital Status']=df['Patient\'s Vital Status'].map(dict)
dict={'M1':0,'M0':1}
df['M Stage']=df['M Stage'].map(dict)

```

```python
#Converting Categorical data using LabelEncoder of Sklearn 
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df['Cancer Type Detailed']=label_encoder.fit_transform(df['Cancer Type Detailed'])
df['Cancer Type Detailed'].unique() # invasice ductal=3,invasive lobular and ductal=6,invasive lobular=4,Invasive Cancer,NOS=2,Adenoid Cystic Breast Cancer=0
                                    # Invasive Mixed Mucinous Carcinoma=5,breast=1,metaplastic breast cancer=8,Invasive breast carcinoma=7
df['Sex']=label_encoder.fit_transform(df['Sex']) #female=0 ,male=1
df['Sex'].unique()                                                          #man=0,unk=4
df['T Stage']=label_encoder.fit_transform(df['T Stage'])
df['T Stage'].unique() 
df['Menopausal Status At Diagnosis']=label_encoder.fit_transform(df['Menopausal Status At Diagnosis']) #pre=3 ,post=2 ,peri=1
df['Menopausal Status At Diagnosis'].unique() 
df['N Stage']=label_encoder.fit_transform(df['N Stage']) 
df['N Stage'].unique()
```

```python
df['ER Status of the Primary']=label_encoder.fit_transform(df['ER Status of the Primary']) 
df['ER Status of the Primary'].unique()
df['HER2 IHC Status Primary']=label_encoder.fit_transform(df['HER2 IHC Status Primary']) 
df['HER2 IHC Status Primary'].unique()
df['Overall Primary Tumor Grade']=label_encoder.fit_transform(df['Overall Primary Tumor Grade']) 
df['Overall Primary Tumor Grade'].unique()
```

```python
df['PR Status of the Primary']=label_encoder.fit_transform(df['PR Status of the Primary']) 
df['PR Status of the Primary'].unique()
df['Stage At Diagnosis']=label_encoder.fit_transform(df['Stage At Diagnosis']) 
df['Stage At Diagnosis'].unique()
```

```python
# Importing train_test_split from Sklearn
from sklearn.model_selection import train_test_split
```

```python
#splitting data into test and train dataset
X_train,X_test,y_train,y_test=train_test_split(df.drop(['Patient\'s Vital Status'],axis=1),df['Patient\'s Vital Status'],test_size=0.25,random_state=42)
```

```python
X_test
X_test.shape
```

```python
# Training the model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
model=RandomForestClassifier(criterion='gini',n_estimators=1000,random_state=42)
model.fit(X_train,y_train)
```

```python
#Predication on test data and calculation accuracy
prediction=model.predict(X_test)
from sklearn import metrics
print('accuracy=',metrics.accuracy_score(y_test,prediction))
```

```python
# Checking the percentage of survival by giving row number into row variable for example-row 588
row=X_test.loc[588]
yhat = model.predict_proba([row]) #Predict_probs() calulates two problabilities,probability of getting 0 will be for nonSurvival
print(yhat)                                  # and probability of getting 1 for survival
#The Random Forest simply votes among the results. The predict_proba() returns the number of votes for each class, divided by 
#the number of trees in the forest
```

```python
# get percentage of survival
yhat=np.array(yhat)
p_survive=yhat[:,1]*100 # First value is for non survival ,second value is for survival
print('data=%s Survival=%.3f%% '% (row,p_survive)) #Gives the percentage of survival with the data
```
